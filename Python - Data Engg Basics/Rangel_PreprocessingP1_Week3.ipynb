```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
#Initial characterization of the dataframe. I use print for the columns and shape 
#to show the ways I like to visualize the initial data.

path = r'D:\Users\MSI\Desktop\Sparta\pythonweek3'
sales = pd.read_csv(path + '\sales.xls')

print(sales.columns)
print(sales.shape)
sales.head()
```

```python
#Removing duplicate row values
df = sales.drop_duplicates() 
print(df.shape)
df
```

```python
df.to_csv(path + '\sales.xls')
```

```python
# Express price in thousands
# We first remove the commas in some of the price values, so we can convert it to a float.

df.Price = df.Price.replace(',','', regex=True)
df['Price'] = df['Price'].astype('float')
```

```python
df.Price = df.Price/1000
df.rename(columns={'Price':'Price_thousands'}, inplace= True)
df['Price_thousands']
```

```python
# Select name, city, state, and country data of product 3 bought by customers from united states and paid
# thru Visa/Amex. Store in empty.csv

df1 = df.loc[(df.Payment_Type == 'Visa') | (df.Payment_Type == 'Amex')]
df1 = df1.loc[df1.Product == 'Product3']
df1 = df1.loc[df1.Country == 'United States']
df1[['Name','City','State','Country']]
df1
```

```python
df1.to_csv(path + '\empty.xls')
```

```python
# Select name, price, city, state and country data of product 2 bought from New York
# and paid thru mastercard
# There are actually 0 entries with this specific set of filters.

df2 = df.loc[(df.Payment_Type == 'Mastercard') & (df.State == 'NY')]
df2 = df2.loc[df.Product == 'Product2']
df2[['Name','Price_thousands','City','State','Country']]
df2
```

```python
df2.to_csv(path + '\empty1.xls')
```
